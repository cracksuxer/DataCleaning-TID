## Correlación de atributos (Usando el coeficiente de correlación de Pearson)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('datos_homeloan.csv')
```

```python
corr_matrix = data.corr(numeric_only = True)
```

```python
print(corr_matrix)
```

Los valores mostrados de la matriz muestran la correlación que existe entre un par de atributos. Fijándonos en la tabla
no vemos una ningun par de variables que estén por encima de un umbral medio (0.8, 0.9), por lo tanto no se eliminarán por
este método ninguno de los atributos de la tabla.

# Gráfica de correlación

```python
sns.heatmap(corr_matrix, annot=True)
```
